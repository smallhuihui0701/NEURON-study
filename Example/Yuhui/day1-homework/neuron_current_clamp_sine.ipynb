```python
from neuron import h
from neuron.units import mV, ms
import plotly.graph_objects as go
import numpy as np

h.load_file("stdrun.hoc")
```

```python
soma = h.Section(name="soma")
h.pas.insert(soma)

ic = h.IClamp(soma(0.5))
ic.delay = 0
ic.dur = 100
```

```python
ramp_t = h.Vector(np.arange(0 * ms, 30 * ms, 0.025 * ms))
ramp_amp = h.Vector(np.sin(ramp_t)) * 10

ramp_amp.play(ic._ref_amp, ramp_t, True) # last True says interpolate
```

```python
t = h.Vector().record(h._ref_t)
v = h.Vector().record(soma(0.5)._ref_v)
amp = h.Vector().record(ic._ref_i)
```

```python
h.finitialize(-70 * mV)
h.continuerun(30 * ms)
```

```python
fig = go.Figure()
```

```python
fig.add_trace(
    go.Scatter(x=t, y=amp)
).show()
```

```python
go.Figure().add_trace(
    go.Scatter(x=t, y=v)
).show()
```
