```python
from neuron import h, gui
from neuron.units import um, mV, ms
import plotly
import numpy as np

h.load_file('stdrun.hoc')
```

```python
axon = h.Section(name='axon')
```

```python
axon.L = 5001 * um
axon.diam = 1 * um
axon.nseg = 1001
```

```python
h.hh.insert(axon)
```

```python
iclamp = h.IClamp(axon(0))
iclamp.delay = 1
iclamp.dur = 0.1
iclamp.amp = 1
```

```python
rvp = h.RangeVarPlot('v', axon(0), axon(1))
```

```python
h.finitialize(-65 * mV)
```

```python
my_plot = rvp.plot(plotly, name="t=0ms", line={"width": 4})
```

```python
my_plot.show()
```

```python
for tstop in np.arange(0.5, 10.5, 0.5):
    h.continuerun(tstop)
    my_plot = rvp.plot(my_plot, name=f"t={tstop}ms", line={"width": 4})
```

```python
my_plot = my_plot.update_layout({
    "xaxis_title": "position [um]",
    "yaxis_title": "membrane potential [mV]",
    })
```

```python
my_plot.show()
```
